# Problem 4

A certain paint can be produced using four different production processes (P1, P2, P3, and P4). The processing cost of each liter in any of the four available processes, the maximum capacity of each process, and the setup costs are given below:

|Process|Setup Cost (\$)|Processing Cost (\$/L)|Capacity (L)|
|--|--|--|--|
|P1|5,000|0.60|20,000|
|P2|6,000|0.50|15,000|
|P3|10,000|0.40|40,000|
|P4|6,000|0.40|25,000|

Assume that a daily demand of 45,000 L must be fulfilled. Formulate the problem as a programming model to determine the daily production schedule in order to minimize total costs.

Resposta:

Assumimos que o Setup Cost é um valor de custo diário

Assumimos que o processo de produção admite a fabricação de frações de litro de tinta

A tinta pode ser produzida por qualquer processo P1-P4

Por dia, é necessário produzir pelo menos 45.000 L


**1 - Definição das varáveis de decisão**

O problema pergunta quantos litros de tinta devem ser produzidos utilizando cada processo a fim de atender a demanda de 45.000 L com o menor custo possível.

$Lp1$ -> Quantidade de litros produzidos pelo processo P1

$Lp2$ -> Quantidade de litros produzidos pelo processo P2

$Lp3$ -> Quantidade de litros produzidos pelo processo P3

$Lp4$ -> Quantidade de litros produzidos pelo processo P4

$Lpn \in R$

**2 - Definição da função objetivo**

O objetivo é minimizar o custo total de produção

$Ct$ -> Custo total

$Cp1$ -> Custo parcial com o processo P1

$Cp2$ -> Custo parcial com o processo P2

$Cp3$ -> Custo parcial com o processo P3

$Cp4$ -> Custo parcial com o processo P4

$Ct = Cp1 + Cp2 + Cp3 + Cp4$

O custo parcial de cada processo é dado por:

$Cpn = SCpn + Lpn \times PCpn$

Em que $SC$ representa o Setup Cost e $PC$ representa o Processing Cost/L

Então:

$Ct = \sum_{n=1}^{4} SCpn + Lpn \times PCpn$


**3 - Definição das restrições**

$s.t.:$

$0 \leq Lp1 \leq 20.000 L$  

$0 \leq Lp2 \leq 15.000 L$  

$0 \leq Lp3 \leq 40.000 L$  

$0 \leq Lp4 \leq 25.000 L$  

$\sum_{n=1}^{4} Lpn \geq 45.000 $



````python
from pyomo.environ import *

# creating a model
model = ConcreteModel()

# problem parameters

SCp1 = 5000
SCp2 = 6000
SCp3 = 10000
SCp4 = 6000

PCp1 = 0.6
PCp2 = 0.5
PCp3 = 0.4
PCp4 = 0.4

# declaring decision variables
model.Lp1 = Var(domain = NonNegativeReals)
model.Lp2 = Var(domain = NonNegativeReals)
model.Lp3 = Var(domain = NonNegativeReals)
model.Lp4 = Var(domain = NonNegativeReals)


# declaring objective

Ct_expression = SCp1 + model.Lp1 * PCp1 + SCp2 + model.Lp2 * PCp2 + SCp3 + model.Lp3 * PCp3 + SCp4 + model.Lp4 * PCp4
model.Ct = Objective(expr = SCp1 + model.Lp1 * PCp1 + SCp2 + model.Lp2 * PCp2 + SCp3 + model.Lp3 * PCp3 + SCp4 + model.Lp4 * PCp4, sense=minimize)


# declaring constraints

model.capacity_P1 = Constraint(expr = model.Lp1 <= 20000)
model.capacity_P2 = Constraint(expr = model.Lp2 <= 15000)
model.capacity_P3 = Constraint(expr = model.Lp3 <= 40000)
model.capacity_P4 = Constraint(expr = model.Lp4 <= 25000)
model.total_production = Constraint(expr = model.Lp1 + model.Lp2 + model.Lp3 + model.Lp4 >= 45000)

model.pprint()
````

````python
solver = SolverFactory('cbc', executable='/usr/bin/cbc')
results = solver.solve(model)
results.write()

# display solution
print('\nCost = ', model.Ct())

print('\nDecision Variables')
print('Lp1 = ', model.Lp1())
print('Lp2 = ', model.Lp2())
print('Lp3 = ', model.Lp3())
print('Lp4 = ', model.Lp4())

print('\nConstraints')
print('P1 capacity  = ', model.capacity_P1())
print('P2 capacity  = ', model.capacity_P2())
print('P3 capacity  = ', model.capacity_P3())
print('P4 capacity  = ', model.capacity_P4())
print('Total Production has to be higher or equal to ', model.total_production())
````
